```python
# lambda function - Row wise operation
def revenue_budget(revenue,budget):
  if revenue > 0:
    new_profit = revenue - budget
  else:
    new_profit = np.nan

  return new_profit
```

```python
def convert_to_year(datevalue):
  return datevalue.year
```

```python
df['Year_of_release_1'] = df['new_release_date'].apply(convert_to_year)
```

```python
df.head(1)
```

```python
# We can find the maximum movie runtime across different years
df.groupby('Year_of_release_1')['runtime'].median().reset_index()
```

```python
df.groupby('Year_of_release_1')['imdb_id'].count().median()
```

```python
df.groupby('Year_of_release_1',as_index=False).agg({"runtime": "sum"})
```

```python
# Group the data frame by month and item and extract a number of stats from each group
df.groupby(['Year_of_release_1', 'original_language'],as_index=False).agg({'runtime':'mean','revenue': "sum",'imdb_id':"count"}).sort_values(by='Year_of_release_1',ascending = True).rename(columns ={'runtime':'avg_runtime','revenue':'total_revenue','imdb_id':'num_movies'})
```
