```python
# lambda function - Row wise operation
def revenue_budget(revenue,budget):
  if revenue > 0:
    new_profit = revenue - budget
  else:
    new_profit = np.nan

  return new_profit
```

```python
df['new_profit'] = df.apply(lambda x: revenue_budget(x['revenue'],x['budget']),axis =1)
```
